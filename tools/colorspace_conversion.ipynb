```python
import numpy as np
from skimage.io import imread
from skimage.color import rgb2yuv, yuv2rgb
import matplotlib.pylab as plt
```

```python
im = imread(r"D:\Test_Result\BasicVSRPP_c64n7_VimeoPretrained_300000iter\011\00000000.png")
im1 = rgb2yuv(im)
im2 = yuv2rgb(im1)
```
